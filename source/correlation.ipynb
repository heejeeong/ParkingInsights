```python
#import
import pandas as pd
import haversine
from openpyxl import load_workbook
```

```python
# 파일 불러오기
enforcement_df = pd.read_excel('../data/filtered_parking_data_low.xlsx')
parking_lot_df = pd.read_excel('../data/불법주정차_low.xlsx')
```

```python
# 해당 구 추출
selected_rows_1 = parking_lot_df[parking_lot_df['주소'].str.contains('도봉구')]
selected_rows_2 = enforcement_df[enforcement_df['주소'].str.contains('도봉구')]

selected_rows_1 = selected_rows_1.reset_index(drop=True)
selected_rows_2 = selected_rows_2.reset_index(drop=True)

#중구 0.81
#종로구 0.68
#서초구 0.91
# 0.8

#양천구 0.91
#중랑구 0.74
#도봉구 0.54
# 0.73
```

```python
# 새로운 엑셀 파일 생성

data = {
    '위도(주차장)': ['','',''],
    '경도(주차장)': ['','',''],
    '위도(불법)': ['','',''],
    '경도(불법)': ['','','']
}

# 데이터프레임 생성
df = pd.DataFrame(data)

# 엑셀 파일로 저장
df.to_excel('../result/newfile6.xlsx', index=False)
```

```python
df = pd.read_excel('../result/newfile6.xlsx')

for i in range(len(selected_rows_1)):
    for j in range(len(selected_rows_2)):
        data1 = selected_rows_1['위도'][i]
        data2 = selected_rows_1['경도'][i]
        data3 = selected_rows_2['위도'][j]
        data4 = selected_rows_2['경도'][j]
        res = haversine.haversine((data1, data2), (data3, data4))

        if res < 0.1:
            new_row = {
                '위도(주차장)': data1,
                '경도(주차장)': data2,
                '위도(불법)': data3,
                '경도(불법)': data4
            }
            df = df.append(new_row, ignore_index=True)

df.to_excel('../result/newfile6.xlsx', index=False)
```
